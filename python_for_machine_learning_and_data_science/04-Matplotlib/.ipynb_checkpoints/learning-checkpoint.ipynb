```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[1].plot(a, b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(a, b)
axes[0, 1].plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

for ax in axes:
    ax.plot(a, b)
    
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 5), dpi=100)

k = np.linspace(0, 50, 100)
j = np.sin(k)

axes[0, 0].plot(x, y)
axes[0, 0].set_title('Linear')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y')

axes[0, 1].plot(x, y)
axes[0, 1].set_title('Linear')

axes[1, 0].plot(k, j)
axes[1, 0].set_xlim(0, 50)
axes[1, 0].set_title('Sin')

axes[1, 1].plot(a, b)
axes[1, 1].set_title('Quadratic')

fig.suptitle('Figure Level', fontsize=20)

plt.tight_layout()
# plt.subplots_adjust(wspace=0.3, hspace=0.5)

# fig.savefig('test.png', bbox_inches='tight')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))

axes.plot(a, b, label='x ^ 2')
axes.plot(x, y ** 3, label='y ^ 3')
axes.set_xlim(0, 10)

plt.legend(loc=(0.5, 0.7))
```

```python
fig = plt.figure()

axes_1 = fig.add_axes([0, 0, 1, 1])
axes_1.plot(x, y ** 3, label='y')
axes_1.plot(a, b, label='b')
axes_1.set_xlim(0, 10)

plt.legend()
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure(figsize=(10, 5), dpi=100)

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, x, color='#938fab', label='x', linewidth=6, ls='--')
ax.plot(x, x + 1, color='#ff3921', label='x + 1', lw=9, linestyle=':', marker='v', ms=15, markerfacecolor='black')

# custom linestyle
ax.plot(x, x + 2, color='green', label='x + 2', lw=6)[0].set_dashes([5, 2, 10, 2])

ax.plot(x, x + x, color='#392fab', label='x + 3', lw=2, ls='--', marker='o', markersize=15, markerfacecolor='red', markeredgecolor='orange', markeredgewidth=4)

plt.legend()
```
