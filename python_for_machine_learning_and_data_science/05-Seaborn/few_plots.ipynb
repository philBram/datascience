```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.info()
```

```python
x = np.linspace(0, 100, 50)
y = x ** 2
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6), dpi=100)
fig.suptitle('Testing matplotlib')

for i in range(len(axes)):
    for j in range(len(axes[i])):
        axes[i, j].plot(x, y, color='r', marker='o', markerfacecolor='b', markeredgewidth=3, markeredgecolor='y', label=f'axes{i, j}')
        axes[i, j].scatter(x, y * 3, c='y', label=f'axes{i, j}', marker='^')
        axes[i, j].set_xlim(0, 100)
        axes[i, j].set_ylim(0, 10000)
        axes[i, j].set_xlabel('x')
        axes[i, j].set_ylabel('y')
        
axes[0, 0].grid()
        
plt.legend()
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(12, 6), dpi=100)

axes = fig.add_axes([0, 0, 1, 1])
axes.barh(['A', 'B', 'C'], [5, 29, 11], color='r', height=0.9)

axes.grid(which='both', axis='y')
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(20, 6), dpi=100)

sns.countplot(data=df, x='level of education', ax=axes[0, 0])
sns.barplot(data=df, x='level of education', y='salary', ax=axes[0, 1])
sns.scatterplot(data=df, x='salary', y='sales', ax=axes[1, 0], hue='level of education', size='level of education', style='level of education')
sns.rugplot(data=df, x='sales', ax=axes[1, 1], height=0.5)
```

```python
sns.set_theme(rc={'figure.figsize': (12, 6), 'figure.dpi': 100})
sns.barplot(data=df, x='level of education', y='salary', palette='inferno', ci='sd')
```
