```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
# plt.figure(figsize=(10, 4), dpi=100)
fig = plt.figure(figsize=(10, 4), dpi=100)
axes_1 = fig.add_axes([0, 0, 1, 1])
sns.boxplot(data=df, y='writing score', x='parental level of education', hue='test preparation course', palette='Dark2', ax=axes_1)
```

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.boxplot(data=df, x='writing score', y='parental level of education', hue='test preparation course', palette='Set1')
```

```python
sns.set_style(style='whitegrid')

fig, (axes_1, axes_2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 6), dpi=100)

sns.boxplot(data=df, y='parental level of education', x='writing score', ax=axes_1, hue='test preparation course')
sns.violinplot(data=df, y='parental level of education', x='writing score', ax=axes_2, hue='test preparation course', split=True)

axes_1.legend(bbox_to_anchor=(0, 0))
axes_2.legend(bbox_to_anchor=(0, 0))
plt.tight_layout()
```

```python
sns.set_style(style='darkgrid')
sns.violinplot(data=df, x='parental level of education', y='writing score', inner='quartile')
```

```python
sns.violinplot(data=df, x='parental level of education', y='writing score', inner=None)
```

```python
sns.violinplot(data=df, x='parental level of education', y='writing score', inner='stick')

plt.tight_layout()
```

```python
sns.violinplot(data=df, x='parental level of education', y='writing score', bw=0.1)

plt.tight_layout()
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.swarmplot(data=df, x='math score', y='gender', size=4, hue='test preparation course')

plt.legend(bbox_to_anchor=(1.2, 1.02))
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.swarmplot(data=df, x='math score', y='gender', size=4, hue='test preparation course')

plt.legend(bbox_to_anchor=(1.2, 1.02))
```

```python
sns.set_style('whitegrid')

g = sns.PairGrid(df, hue='gender', palette='Dark2')

g = g.map_upper(sns.scatterplot, marker='x')
g = g.map_diag(sns.kdeplot, shade=True)
g = g.map_lower(sns.kdeplot)

g = g.add_legend(bbox_to_anchor=(1, 1))
```
