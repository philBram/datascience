```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def draw_learning_rate(weights_slope, weights_intercept, weights_cost, plt_type='3d', ax=None, size=30):

    if plt_type == '3d':

        if ax == None:
            ax = plt.axes(projection='3d')
            ax.set_xlabel('slope')
            ax.set_ylabel('intercept')
            ax.set_zlabel('cost')
            ax.set_title('learning rate')

        for i in range(len(weights_cost) - 1):
            ax.plot([weights_slope[i], weights_slope[i + 1]], [weights_intercept[i], weights_intercept[i + 1]], zs=[weights_cost[i], weights_cost[i + 1]], c='b')
            
        ax.scatter(weights_slope, weights_intercept, weights_cost, c='r', s=size)
```

```python
def function_z(x, y):
    return (x ** 2 + y ** 2)
```

```python
slope = np.linspace(-5, 5, 10)
intercept = np.linspace(-5, 5, 10)
cost = np.linspace(-5, 50, 10)

N = 50

x_values = np.linspace(-5, 5, N)
y_values = np.linspace(-5, 5, N)

X, Y = np.meshgrid(x_values, y_values)
Z = function_z(X, Y)
```

```python
%matplotlib widget
fig = plt.figure(figsize=(10, 5), dpi=100)

ax2 = plt.subplot(121, projection='3d')
ax2.scatter(X, Y, Z)
ax2.plot_wireframe(X, Y, Z)

ax1 = plt.subplot(122, projection='3d')
ax1.set_title('test')

draw_learning_rate(slope, intercept, cost, ax=ax1)
```
